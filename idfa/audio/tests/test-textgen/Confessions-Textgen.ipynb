```python
from textgenrnn import textgenrnn

textgen = textgenrnn()
```

```python
textgen.train_from_file('confessions.txt', num_epochs=5)
```

```python
with open('gan_confessions.txt', 'a') as fd:
    for i in range(100):                
        confession = textgen.generate(return_as_list=True)
        fd.write(confession[0]  + '\n')
```
